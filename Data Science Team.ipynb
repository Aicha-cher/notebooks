## Let's find the word that appears the most often in our corpus...

```python
from collections import Counter
import io
import sys

cnt = Counter()

with io.open('C:/Users/c-bourguignat/Documents/perso/14-ML meetup/train.csv', 'r', encoding='cp1252') as f:
    for line in f:
        for word in line.split():
            if len(word) > 3:
                cnt[word.lower()] += 1

total_word_count = sum(cnt.values())
for word, count in cnt.most_common(1):
    print count, float(count) / float(total_word_count), word
```

## Now do TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize

class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc) if len(t) > 2]

traindata = pd.read_csv('C:/Users/c-bourguignat/Documents/perso/14-ML meetup/train.csv')

X = traindata['text']
y = traindata['target']

vect = TfidfVectorizer(stop_words='english',encoding='cp1252',min_df=3,max_df=1.0,strip_accents='unicode',analyzer='word',
                       ngram_range=(1,2),use_idf=1,smooth_idf=1,sublinear_tf=1, tokenizer=LemmaTokenizer())

Xtfidf = np.array(vect.fit_transform(X).todense())
featcolstfidf = vect.get_feature_names()
print Xtfidf.shape
```

## Find out wich terms are most typical of each class

## .... Now build a machine learning model

#### Train the model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200, max_depth=15)
model.fit(Xtfidf, y)
```

#### Find out the most important terms used by the model to discriminate classes

```python
idx = np.argsort(model.feature_importances_)
for i in range(10):
    print featcolstfidf[idx[-i-1]], " : ", model.feature_importances_[idx[-i-1]]
```
